```python
import re           
import nltk            
import string    
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords              
import matplotlib.pyplot as plt 
```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

```python
text="Hello hello I like the machine learning very much! is the and excited to learn?"
text
```

```python
tokens=word_tokenize(text.lower())
tokens
```

```python
stop_words=set(stopwords.words('english'))
#print(stop_words)
tokens_clean=[word for word in tokens if word not in stop_words and word not in string.punctuation]
print(tokens_clean)
```

```python
word_freq=Counter(tokens_clean)
word_freq
```

```python
top_words=word_freq.most_common(10)
words,counts = zip(*top_words)
```

```python
plt.figure(figsize=(5,2))
plt.bar(words,counts,color="skyblue")
plt.xlabel("words")
plt.ylabel("freq")
plt.show()
```
