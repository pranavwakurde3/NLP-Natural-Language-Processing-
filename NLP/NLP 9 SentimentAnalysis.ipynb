```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download("vader_lexicon")
```

```python
senti=SentimentIntensityAnalyzer()
senti
```

```python
texts=["I love this Product!",
      "This is my worst movie ever.",
      "It's okey,not great though.",
      "I hate this so much."]
```

```python
for text in texts:
    scores=senti.polarity_scores(text)
    print("\nText:",text)
    print("Sentiment Score:",scores)
```

```python
compound=scores['compound']
if compound>=0.05:
    sentiment="Positive"
elif compound<=-0.05:
    sentiment="Negative"
else:
    sentiment="Neutral"

print("Compound Score:",compound)
print("Predicted Sentiment:",sentiment)
```
