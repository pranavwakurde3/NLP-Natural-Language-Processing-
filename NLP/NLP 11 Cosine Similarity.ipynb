```python
# Step 1: Import libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Step 2: Create sample texts
text1 = "Artificial intelligence is the future of technology."
text2 = "Machine learning and artificial intelligence are related fields of technology."
```

```python
# Step 3: Convert text to TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([text1, text2])
```

```python
# Step 4: Compute cosine similarity
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
```

```python
# Step 5: Print result
print("Cosine Similarity between the two texts:", similarity[0][0])
```
