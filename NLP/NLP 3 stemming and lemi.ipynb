```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import PorterStemmer
from nltk import WordNetLemmatizer
```

```python
nltk.download("wordnet")
```

```python
text="Hello hello I like the machine learning very much! is the and excited to  running learn?"
text
```

```python
tokens=word_tokenize(text.lower())
```

```python
ps=PorterStemmer()
stemmer=[ps.stem(word) for word in tokens]
stemmer
```

```python
wnl=WordNetLemmatizer()
lemmatizer=[wnl.lemmatize(word) for word in tokens]
lemmatizer
```
