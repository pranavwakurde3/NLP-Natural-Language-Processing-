```python
x="my string is 230GAIB91_05"
y="my string is 230GAIB92_06"
```

```python
num=re.findall('\d',x)
string=re.findall('[a-zA-Z]',x)
special_char=re.findall('[^a-zA-Z0-9]',x)
```

```python
print("Num:",num)
print("String:",string)
print("Special_char:",special_char)
```
