```python
import spacy
import nltk
from nltk.tokenize import word_tokenize
```

```python
text="Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne on April 1, 1976, in Cupertino, California. The company's first product was the Apple I, a personal computer designed by Wozniak.In 2007, Apple introduced the iPhone, a revolutionary smartphone that transformed the mobile industry.Tim Cook became CEO in 2011, succeeding Steve Jobs. Apple is known for its innovative products, including the Mac, iPad, Apple Watch, and AirPods. The company's headquarters, Apple Park, is located in Cupertino and was designed by Foster + Partners. As of October 2025, Apple is one of the most valuable companies in the world, with a significant presence in technology and entertainment."
```

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp(text)
```

```python
for ent in doc.ents:
    print("Word:",ent.text)
    print("NER:",ent.label_)
```
