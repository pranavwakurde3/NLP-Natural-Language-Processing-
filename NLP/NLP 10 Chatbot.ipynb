```python
rules=[(r'hi|hello|hey', ['Hello!', 'Hi there!', 'Greetings!']),
    (r'how are you', ["I'm doing well, thank you!", "I'm a bot, but I'm functioning perfectly."]),
    (r'what is your name', ["I don't have a name.", "You can call me Chatbot."]),
    (r'bye|goodbye', ['Goodbye!', 'See you later!', 'Farewell!']),
    (r'thank you|thanks', ['You\'re welcome!', 'No problem!', 'Glad to help.']),
    (r'.*', ["I'm not sure how to respond to that.", "Can you rephrase your question?", "Please tell me more."])
]
```

```python
class RuleBasedChatbot:
    def __init__(self,rules):
        self.chat=Chat(rules,reflections)
    def respond(self,user_input):
        return self.chat.respond(user_input)
```
