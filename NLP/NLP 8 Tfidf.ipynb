```python
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
d0="Geeks for geeks"
d1="Geeks"
d2="P2j"
string=[d0,d1,d2]
string
```

```python
tfidf=TfidfVectorizer()
result=tfidf.fit_transform(string)
result
```

```python
print(tfidf.get_feature_names_out())
```

```python
print(tfidf.idf_)
```

```python
print("Word indes:",tfidf.vocabulary_)
```

```python
print("tfidf values:",result)
```

```python
print("tfidf in matrix form:",result.toarray())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tfidf_dense=result.toarray()
df=pd.DataFrame(tfidf_dense,columns=tfidf.get_feature_names_out())
plt.figure(figsize=(8,4))
sns.heatmap(df,annot=True,cmap="YlGnBu",cbar=True)
plt.title("TF-IDF Heatmap")
```
