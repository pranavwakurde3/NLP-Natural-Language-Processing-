```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
```

```python
corpus=["Machine leaarning is fun and powerful",
        "Deep learning is a subset og machine learning",
        "Artificial Intelligence and machine learning are closely related",
        "Python is great programming language for ai & ml student"]
```

```python
tokenized_corpus=[simple_preprocess(sentence) for sentence in corpus]
```

```python
model=Word2Vec(sentences=tokenized_corpus,
               vector_size=100,
               window=5,
               min_count=1,
               sg=1,
               epochs=10)
```

```python
print("Most similar to 'learning':")
print(model.wv.most_similar('learning',topn=5))
```

```python
similarity_score=model.wv.similarity('machine','learning')
print("Similarity between 'machine' and 'learning':",similarity_score)
```
