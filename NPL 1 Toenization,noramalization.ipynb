```python
#library import
import nltk
import re
from nltk.tokenize import word_tokenize
```

```python
#download pre-trained model
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
text="Hello I like the machine learning very much! and excited to learn?"
text
```

```python
#Convert Upper text to Lower text
text_lower=text.lower()
print("Lower Case:")
print(text_lower)
```

```python
text_no_punct=re.sub(r'[^\w\s]',"",text_lower)
print("\nText without Punct")
print(text_no_punct)
```

```python
#Tokenization - Splitting the words
tokens=word_tokenize(text_no_punct)
print("\nTokenized text:")
print(tokens)
```
